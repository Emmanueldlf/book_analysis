### Using regex

```python
import re
```

```python
pattern = re.compile("Chapter [0-9+]")
result = re.findall(pattern, book)
len(result)
```

```python
re.compile(r'Chapter [0-9]', re.UNICODE)
```

## Extract sentences where love appear

```python
pattern = re.compile("[A-Z]{1}[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
findings = re.findall(pattern, book)
findings
```

### Identify the most common words

```python
pattern = re.compile("[a-zA-Z]+")
findings = re.findall(pattern, book.lower())
findings[:5]
```

```python
words ={}
for word in findings:
    if word in words.keys():
        words[word] = words[word] + 1
    else:
        words[word] = 1
```

```python
words_list = [{value, key} for (key, value) in words.items()]
sorted(words_list, reverse=True)
```
